```python
#importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# importing csv
df = pd.read_csv('GrowLocations.csv')
```

```python
#viewing  the data
df
```

 Things to look out for in the data
 1. Some location values are way outside the allowed values for latitude and Longitude.
 2. Some locations are not on the map provided.
 3. The labels of the columns have not be verified so may be incorrect.

```python
# extracting the required data
latlong= df[['Latitude','Longitude']]
```

```python
latlong
```

```python
#renaming the columns switch the lat and long are switched
latlong.rename(columns={'Latitude': 'Long', 'Longitude': 'Lat'}, inplace=True)
```

```python
latlong
```

```python
# viewing the datatypes
latlong.describe
```

```python
#latlong=latlong.drop_duplicates()
```

```python
# looking for nulls
latlong.isna().sum()
```

```python
# finding outliers
latlong.boxplot(['Lat','Long'])
```

```python
# using zscore to remove all values outside of 3 standard deviations
latlong=latlong[(np.abs(stats.zscore(latlong)) < 3).all(axis=1)]
```

```python
latlong.describe
```

```python
# visualizing after removing outliers
latlong.boxplot(['Lat','Long'])
```

The bounding box for the map is as follows:
1. Longitude Min -10.592
2. Longitude Max 1.6848
3. Latitude Min 50.681
4. Latitude Max 57.985

```python
# defining the bounding box of the images
BBox = [-10.592,1.6848,50.681,57.985]
```

```python
# ignoring values outside of the bounding box.
latlong =  latlong[(latlong['Lat'] < BBox[3]) & (latlong['Lat'] > BBox[2])]
latlong =  latlong[(latlong['Long'] < BBox[1]) & (latlong['Long'] > BBox[0])]
```

```python
latlong
```

```python
latlong.describe
```

```python
# importing the image
gbmap = plt.imread('map7.png')
```

```python
# plotting the points on the image
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(latlong.Long,latlong.Lat, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Plotting Grow Data on UK Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(gbmap, zorder=0, extent = BBox, aspect= 'equal')
```
